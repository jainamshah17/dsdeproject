```python
### Imports
import os
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

# Constants
DATA_PATH = "./raw/"
```

# Data Preparation

```python
# Load dataset
files = []
for file in os.listdir(DATA_PATH):
    if file.endswith(".csv"):
        files.append(file)

dfs = []
for file in files:
    d = pd.read_csv(DATA_PATH + file)
    d["region"] = file[0:2]
    dfs.append(d)

df = pd.concat(dfs, axis = 0)
df.head()
```

```python
# Handling null / missing values
df.isna().sum()
```

```python
df = df.drop(columns = ["description"])
df.isna().sum()
```

We have dropped the description column due to null values. Description column requires natural language processing (NLP) related models for analysis, which is out of the scope of our project. We focus more on analysing statistical features like number of views, trending dates, number of likes & comments, etc. on the youtube videos.
